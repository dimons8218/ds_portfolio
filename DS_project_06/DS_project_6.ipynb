**Описание проекта:** 
Вы работаете в IT-компании, которая выполняет на заказ проекты по машинному обучению. К вам обратился фермер, владелец молочного хозяйства «Вольный луг». Он хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ «ЭкоФерма».
Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже. А продавцы и технологи так и норовят приукрасить своих коровок! «ЭкоФерма» готова предоставить подробные данные о своих коровах.

Данные состоят из трёх датасетов:
- ferma_main.csv;
- ferma_dad.csv;
- cow_buy.csv.


**Файл ferma_main.csv** содержит данные о стаде фермера на текущий момент.

*Описание данных:*
- id — уникальный идентификатор коровы.
- Удой, кг — масса молока, которую корова даёт в год (в килограммах).
- ЭКЕ (Энергетическая кормовая единица) — показатель питательности корма коровы.
- Сырой протеин, г — содержание сырого протеина в корме (в граммах).
- СПО (Сахаро-протеиновое соотношение) — отношение сахара к протеину в корме коровы.
- Порода — порода коровы.
- Тип пастбища — ландшафт лугов, на которых паслась корова.
- порода папы_быка — порода папы коровы.
- Жирность,% — содержание жиров в молоке (в процентах).
- Белок,% — содержание белков в молоке (в процентах).
- Вкус молока — оценка вкуса по личным критериям фермера, бинарный признак (вкусно, не вкусно).
- Возраст — возраст коровы, бинарный признак (менее_2_лет, более_2_лет).

Признаки в датасете ferma_main.csv можно разделить на группы.
- Характеристики коровы: id, Порода, порода папы_быка, Возраст.
- Характеристики корма: ЭКЕ (Энергетическая кормовая единица), Сырой протеин, г, СПО (Сахаро-протеиновое соотношение).
- Характеристика пастбища: Тип пастбища.
- Характеристики молока: Удой, кг, Жирность,%, Белок,%, Вкус молока.

**Файл ferma_dad.csv** хранит имя папы каждой коровы в стаде фермера.

*Описание данных:*
- id — уникальный идентификатор коровы.

**Файл cow_buy.csv** — это данные о коровах «ЭкоФермы», которых фермер хочет изучить перед покупкой.

- Текущий_уровень_белков,% — содержание белков в молоке (в процентах).
- Возраст — возраст коровы, бинарный признак (менее_2_лет, более_2_лет).
- Данные в файлах ferma_main.csv и cow_buy.csv похожи, но у датасета cow_buy.csv несколько особенностей.
- Содержание белков и жиров в молоке указано на момент продажи — когда «ЭкоФерма» давала коровам свой корм.

Изучим взаимосвязь между признаками: рассчитаем коэффициенты корреляции Пирсона между признаками и построим матрицу корреляции.

```python
ferma_main.corr()
```

```python
plt.figure(figsize=(10, 6));

sns.heatmap(ferma_main.corr(), annot=True, fmt='.2f');
```

Светлые тона говорят о сильной линейной зависимости, а тёмные — о слабой.
```python
plt.figure(figsize=(10, 6));

sns.heatmap(ferma_main.phik_matrix(), annot=True, fmt='.2f');
```

Теперь видно, что высокая степень корреляции наблюдается между целевым признаком milk_yield_kg и входными признаками efu_energy_feed_unit, stp_sugar_to_protein и fat_content_percent, а при этом заметная связь существует и между этими входными признаками. ## Обучение модели линейной регрессии

Нам предстоит обучить три модели простой линейной регрессии.

Подготовим данные датасета ferma_main.csv и обучим на них первую модель LinearRegression. Используем в качестве целевого признак Удой, кг.

Отберём признаки, которые подходят для модели линейной регрессии. 

На основании корреляционного анализа, можно сказать, что для модели линейной регрессии подходят признаки ЭКЕ и СПО, либо их модификации, учитывающие нелинейный характер связи с удоем.

При этом вне зависимости от наличия линейной связи между признаками на удой коров могут влиять характеристики коровы: (breed, dad_bull_breed, age), характеристики корма (efu_energy_feed_unit, raw_protein_g, stp_sugar_to_protein) и характеристики пастбища (pasture_type). Так, характеристики самой коровы заранее определяют диапазон значений удоя, который может давать корова. Характеристики корма и пастбища влияют на метаболизм коровы и сужают этот диапазон.

Теперь создадим первую модель:
- разобьём данные на тренировочную и тестовую выборки;
- кодируем категориальные признаки с помощью OneHotEncoder;
- масштабируем количественные признаки с помощью StandardScaler;
- обучим модель линейной регрессии на подготовленных данных;
- оценим качество модели на тестовой выборке: рассчитаем R², mSE, RMSE и MAE; проанализируем остатки с помощью графиков
- оформим выводы.

```python
# создадим константу RANDOM_STATE
RANDOM_STATE = 42
```

```python
# сохраним входные (все, кроме целевых, т.е. удоя и вкуса молока), и кроме содержания жиров и белков в молоке)
# и целевые признаки (в данном случае удой)
X_1 = ferma_main.drop(['milk_yield_kg', 'milk_taste', 'fat_content_percent', 'protein_percent' ], axis=1)
y_1 = ferma_main['milk_yield_kg']
# разделим данные на тренировочную и тестовую выборки
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(
    X_1, 
    y_1, 
    random_state=RANDOM_STATE)
# создадим список всех категориальных признаков
cat_col_names_1 = ['breed', 
                   'pasture_type',
                   'dad_bull_breed',
                   'age']
# создадим список всех количественных признаков
num_col_names_1 = ['efu_energy_feed_unit',
                   'raw_protein_g',
                   'stp_sugar_to_protein']

# подготовим признаки (масштабирование и кодирование)

encoder_1 = OneHotEncoder(drop='first', sparse=False)
X_train_ohe_1 = encoder_1.fit_transform(X_train_1[cat_col_names_1])
X_test_ohe_1 = encoder_1.transform(X_test_1[cat_col_names_1])

encoder_col_names_1 = encoder_1.get_feature_names()

scaler_1 = StandardScaler()
X_train_scaled_1 = scaler_1.fit_transform(X_train_1[num_col_names_1])
X_test_scaled_1 = scaler_1.transform(X_test_1[num_col_names_1])

X_train_ohe_1 = pd.DataFrame(X_train_ohe_1, columns=encoder_col_names_1)
X_test_ohe_1 = pd.DataFrame(X_test_ohe_1, columns=encoder_col_names_1)

X_train_scaled_1 = pd.DataFrame(X_train_scaled_1, columns=num_col_names_1)
X_test_scaled_1 = pd.DataFrame(X_test_scaled_1, columns=num_col_names_1)

X_train_1 = pd.concat([X_train_ohe_1, X_train_scaled_1], axis=1)
X_test_1 = pd.concat([X_test_ohe_1, X_test_scaled_1], axis=1)

# обучим модель и получим предсказания на тестовых данных
model_lr_1 = LinearRegression()
model_lr_1.fit(X_train_1, y_train_1)
predictions_1 = model_lr_1.predict(X_test_1)

# рассчитаем значения R2, NSE, RMSE и MAE для линейной регрессии
r2_lr_1 = r2_score(y_test_1, predictions_1)
mse_1 = mean_squared_error(y_test_1, predictions_1)
rmse_1 = mean_squared_error(y_test_1, predictions_1, squared=False)
mae_1 = mean_absolute_error(y_test_1, predictions_1)

# выведем полученные значения R2, NSE, RMSE и MAE
print("R2 LinearRegression 1 =", r2_lr_1)
print("MSE 1 =", mse_1)
print("RMSE 1 =", rmse_1)
print("MAE 1 =", mae_1)

# проведём анализ остатков перед этим рассчитаем остатки
residuals_1 = y_test_1 - predictions_1

# построим графики по количественным признакам — гистограмму частотности распределения и диаграмму рассеяния
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(x=residuals_1)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(x=predictions_1, y=residuals_1)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
```

Выводы:
- R² модели составил 0.784, что довольно близко к 1, а значит, модель работает достаточно точно;
- анализ остатков показал, что они распределены нормально и график симметричен относительно нуля, что означает, что среднее значение остатков модели равно 0. Диаграмма рассеяния остатков показывает, что их дисперсия непостоянна, что говорит об их нелинейном распределении и о необходимости введения дополнительных признаков, значит, требуются дополнительные исследования.

Подготовим данные с учётом связи входных признаков с целевым и обучим на них вторую модель LinearRegression.

Посмотрим на диаграммы рассеяния: на них можно увидеть, что некоторые входные признаки связаны с целевым нелинейно. Исключим нелинейность, чтобы эти признаки были информативны для линейной регрессии. Вот несколько способов, которыми это можно сделать:
- СПО и Удой связаны нелинейно: на их диаграмме рассеяния наблюдения сгруппированы в два кластера, причём есть такое значение СПО, которое можно считать границей между кластерами. Чтобы устранить нелинейность, преобразуем СПО в категориальный бинарный признак, использовав эту границу;
- ЭКЕ и Удой также связаны нелинейно, это нелинейность можно исключить, если ввести новый признак ЭКЕ в квадрате.

Устраним сначала нелинейную связь между удоем и СПО. Создадим функцию, которая будет присваивать значению СПО ниже 0.92 категорию 0, а в остальных случаях - категорию 1:

```python
def stp_cat(stp):
    if stp < 0.92: # если СПО < 0.92, удой < 6000 кг, что не подходит фермеру:
        return 0
    else:
        return 1
```

Для второй модели создадим датафрейм ferma_main_2, куда передадим датафрейм ferma_main без столбцов 'efu_energy_feed_unit' и 'stp_sugar_to_protein'. Вместо них создадим столбцы 'stp_cat' и 'efu_square'. Также не будем добавлять в новый датафрейм столбцы 'fat_content_percent' и 'protein_percent'.

Столбец 'stp_cat' получит значения в результате подставления данных из столбца 'stp_sugar_to_protein' датафрейма ferma_main в функцию stp_cat:

```python
ferma_main_2 = ferma_main.drop(['efu_energy_feed_unit',
                                'stp_sugar_to_protein', 
                                'fat_content_percent', 
                                'protein_percent'
                               ], axis=1)

ferma_main_2['stp_cat'] = ferma_main['stp_sugar_to_protein'].apply(stp_cat)
    
ferma_main_2[['milk_yield_kg','stp_cat']].sample(5)
```

```python
# сохраним входные (все, кроме целевых, т.е. удоя и вкуса молока) и целевые признаки (в данном случае удой)
X_3 = ferma_main_3.drop(['milk_yield_kg', 'milk_taste'], axis=1)
y_3 = ferma_main_3['milk_yield_kg']
# разделим данные на тренировочную и тестовую выборки
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(
    X_3, 
    y_3, 
    random_state=RANDOM_STATE)
# создадим список всех категориальных признаков
cat_col_names_3 = ['breed', 
                   'pasture_type',
                   'dad_bull_breed', 
                   'age',
                   'stp_cat',
                   'dad_name']
# создадим список всех количественных признаков
num_col_names_3 = ['efu_square',
                   'raw_protein_g']

# подготовим признаки (масштабирование и кодирование)

encoder_3 = OneHotEncoder(drop='first', sparse=False)
X_train_ohe_3 = encoder_3.fit_transform(X_train_3[cat_col_names_3])
X_test_ohe_3 = encoder_3.transform(X_test_3[cat_col_names_3])

encoder_col_names_3 = encoder_3.get_feature_names()

scaler_3 = StandardScaler()
X_train_scaled_3 = scaler_3.fit_transform(X_train_3[num_col_names_3])
X_test_scaled_3 = scaler_3.transform(X_test_3[num_col_names_3])

X_train_ohe_3 = pd.DataFrame(X_train_ohe_3, columns=encoder_col_names_3)
X_test_ohe_3 = pd.DataFrame(X_test_ohe_3, columns=encoder_col_names_3)

X_train_scaled_3 = pd.DataFrame(X_train_scaled_3, columns=num_col_names_3)
X_test_scaled_3 = pd.DataFrame(X_test_scaled_3, columns=num_col_names_3)

X_train_3 = pd.concat([X_train_ohe_3, X_train_scaled_3], axis=1)
X_test_3 = pd.concat([X_test_ohe_3, X_test_scaled_3], axis=1)

# обучим модель и получим предсказания на тестовых данных
model_lr_3  = LinearRegression()
model_lr_3.fit(X_train_3, y_train_3)
predictions_3 = model_lr_3.predict(X_test_3)

# рассчитаем значения R2, NSE, RMSE и MAE для линейной регрессии
r2_lr_3 = r2_score(y_test_3, predictions_3)
mse_3 = mean_squared_error(y_test_3, predictions_3)
rmse_3 = mean_squared_error(y_test_3, predictions_3, squared=False)
mae_3 = mean_absolute_error(y_test_3, predictions_3)

# выведем полученные значения R2, NSE, RMSE и MAE
print("R2 LinearRegression 3 =", r2_lr_3)
print("MSE 3 =", mse_3)
print("RMSE 3 =", rmse_3)
print("MAE 3 =", mae_3)

# проведём анализ остатков, перед этим рассчитаем остатки
residuals_3 = y_test_3 - predictions_3

# построим графики по количественным признакам — гистограмму частотности распределения и диаграмму рассеяния
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(x=residuals_3)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(x=predictions_3, y=residuals_3)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
```

Выводы:

- анализ остатков показал, что они распределены нормально и график симметричен относительно нуля, однако диаграмма рассеяния остатков по-прежнему имеет непостоянный вид, что говорит об их нелинейном распределении и о необходимости введения дополнительных признаков для улучшения предсказательной способности модели.

## Обучение модели логистической регрессии

Подготовим данные датасета ferma_main.csv и обучим на них модель LogisticRegression:
- используем в качестве целевого признак Вкус молока;
- при необходимости преобразуем имеющиеся или добавим новые признаки;
- разобьём данные на тренировочную и тестовую выборки;
- кодируем категориальные признаки с помощью OneHotEncoder;
- масштабируем количественные признаки с помощью StandardScaler;
- обучим модель логистической регрессии на подготовленных данных;
- оценим качество модели на тестовой выборке: рассчитаем метрики accuracy, recall, precision;
- построим матрицу ошибок;
- проанализируем модель логистической регрессии.

Создадим функцию milk_taste_cat, которая будет давать вкусному молоку значение 1, а невкусному - 0:

```python
def milk_taste_cat(milk_taste):
    if milk_taste == "вкусно": 
        return 1
    else:
        return 0
```

Создадим датафрейм ferma_main_taste из датафрейма ferma_main_3, куда к столбцу 'milk_taste' применим созданную функцию, и выведем новый датафрейм:

```python
ferma_main_taste = ferma_main_3
ferma_main_taste['milk_taste'] = ferma_main_taste['milk_taste'].apply(milk_taste_cat)
ferma_main_taste['milk_taste']
```

Перейдём к созданию модели:

```python
# сохраним входные (все, кроме целевых, т.е. удоя и вкуса молока),
# и целевые признаки (в данном случае вкус молока))
X_taste = ferma_main_taste.drop(['milk_yield_kg', 'milk_taste'], axis=1)
y_taste = ferma_main_taste['milk_taste']
# разделим данные на тренировочную и тестовую выборки
X_train_taste, X_test_taste, y_train_taste, y_test_taste = train_test_split(
    X_taste, 
    y_taste, 
    random_state=RANDOM_STATE)
# создадим список всех категориальных признаков
cat_col_names_taste = ['breed',
                       'pasture_type',
                       'dad_bull_breed', 
                       'age',
                       'stp_cat',
                       'dad_name']
# создадим список всех количественных признаков
num_col_names_taste = ['efu_square',
                       'raw_protein_g']

# подготовим признаки (масштабирование и кодирование)

encoder_taste = OneHotEncoder(drop='first', sparse=False)
X_train_ohe_taste = encoder_taste.fit_transform(X_train_taste[cat_col_names_taste])
X_test_ohe_taste = encoder_taste.transform(X_test_taste[cat_col_names_taste])

encoder_col_names_taste = encoder_taste.get_feature_names()

scaler_taste = StandardScaler()
X_train_scaled_taste = scaler_taste.fit_transform(X_train_taste[num_col_names_taste])
X_test_scaled_taste = scaler_taste.transform(X_test_taste[num_col_names_taste])

X_train_ohe_taste = pd.DataFrame(X_train_ohe_taste, columns=encoder_col_names_taste)
X_test_ohe_taste = pd.DataFrame(X_test_ohe_taste, columns=encoder_col_names_taste)

X_train_scaled_taste = pd.DataFrame(X_train_scaled_taste, columns=num_col_names_taste)
X_test_scaled_taste = pd.DataFrame(X_test_scaled_taste, columns=num_col_names_taste)

X_train_taste = pd.concat([X_train_ohe_taste, X_train_scaled_taste], axis=1)
X_test_taste = pd.concat([X_test_ohe_taste, X_test_scaled_taste], axis=1)

# обучим модель и получим предсказания на тестовых данных
clf = LogisticRegression()

clf = clf.fit(X_train_taste, y_train_taste)
y_pred_taste = clf.predict(X_test_taste)

y_pred_taste = pd.DataFrame(data=y_pred_taste, columns=['milk_taste'])
print(y_pred_taste.value_counts())

# рассчитаем значения accuracy, precision и recall для логистической регрессии
acc = accuracy_score(y_test_taste, y_pred_taste)
precision = precision_score(y_test_taste, y_pred_taste)
recall = recall_score(y_test_taste, y_pred_taste)

print('accuracy: ', round(acc, 3))
print('precision: ',round(precision,3))
print('recall: ',round(recall ,3))

# построим матрицу ошибок и сохраните её в переменную
cm = confusion_matrix(y_test_taste, y_pred_taste)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.ylabel('True label')
plt.xlabel('Predicted');
```

Выводы:
- модель точно предсказывает результат почти в 62 % случаев;
- на основании матрицы ошибок можно сказать, что в 48 из 157 случаев модель совершает ошибки первого рода, т.е. ложно определяет невкусное молоко как вкусное, а в 12 из 157 случаев - ошибки второго рода, т.е. определяет вкусное молоко как невкусное, поэтому precision меньше recall.

Поскольку главное для заказчика - вкусное молоко, при покупке коров критичнее всего ошибка первого рода, что объясняется необходимостью отбора коров только со вкусным молоком, т.е. минимизацией ошибок False Positive, а значит, более важна метрика precision.

Сведём критичную ошибку к нулю: для этого изменим порог принадлежности к классам и ещё раз рассчитаем значения метрик recall и precision.

```python
# получим прогнозные метки классов и выберем оценки только для класса "1"
y_proba = clf.predict_proba(X_test_taste)[:,1]
data = pd.DataFrame(zip(y_test_taste, y_proba),
columns = ['y_valid', 'y_proba']).sort_values(by='y_proba',ascending=False)

# посчитаем количество ошибок первого рода при разных порогах, для этого создадим функцию
def milk_taste(y_valid,y_pred_taste):
    if y_valid < y_pred_taste:
        return 1
    else:
        return 0

# добавим пороги с помощью функции linspace
thresholds = [round(i,2) for i in np.linspace(0.1,1,num = 20,endpoint=False)]

# применим её к прогнозным значениям при разных порогах
values = []
columns = []
precisions =[]
for i in thresholds:
    columns.append('y_pred_'+str(i))
    data['y_pred_taste'] = data['y_proba'].apply(lambda x: 1 if x>= i else 0)
    data['check'] = data.apply(lambda x : milk_taste(x.y_valid,x.y_pred_taste), 1)
    values.append(data['check'].sum())
    precisions.append(precision_score(data['y_valid'], data['y_pred_taste'], zero_division=0))

taste = pd.DataFrame(zip(columns, values, precisions),columns = ['Порог',
                                                                 'Количество ошибок FP',
                                                                 'Precision']
                    )

taste
```

Видно, что при пороге в 0.73 количество ошибок первого рода находится на уровне 6, а precision равен 0.83. Выберем порог в 0.73 и посмотрим как изменятся метрики, при работе модели на тестовых данных:

```python
y_pred_taste_073 = (clf.predict_proba(X_test_taste)[:, 1] >= 0.73).astype(int)

y_pred_taste_073 = pd.DataFrame(data=y_pred_taste_073, columns=['milk_taste'])

# рассчитаем значения accuracy, precision и recall для логистической регрессии
acc_073= accuracy_score(y_test_taste, y_pred_taste_073)
precision_073 = precision_score(y_test_taste, y_pred_taste_073)
recall_073 = recall_score(y_test_taste, y_pred_taste_073)

print('accuracy 0.73: ', round(acc_073, 3))
print('precision 0.73: ',round(precision_073,3))
print('recall 0.73: ',round(recall_073,3))

# построим матрицу ошибок и сохраним её в переменную
cm_073 = confusion_matrix(y_test_taste, y_pred_taste_073)
sns.heatmap(cm_073, annot=True, fmt='d', cmap='Blues_r')
plt.ylabel('True label')
plt.xlabel('Predicted');
```

Выводы:
- accuracy немного снизился, до 59.2 %;
- precision повысился до 0.833, значительно приблизившись к 1, а, recall, соответственно, снизился примерно в 2.5 раза: с 0.864 до 0.341.
- исходя из анализа модели необходимо повышать метрики accuracy и recall, при сохранении precision на максимально возможном уровне, это позволит верно отбирать большее количество коров с вкусным молоком.
